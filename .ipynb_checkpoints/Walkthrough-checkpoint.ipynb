## From sound to bits

As some of you previously experienced with images (and other forms of data), the data scientist will usually have to process his data and put all features on the same scale. In the case of images, we usually regard them as multidimensional arrays with red, green and blue channels between the values of 0-255.

We will attempt to do the same but with the sound. But sound is a wave - how do we transform a wave of someone saying a digit into something readable by a computer?

To transform this sound wave into something the computer can read (numbers), we can just record the height of this wave at equidistant intervals

This is what is commonly known as sampling - we are taking a reading thousands of times a second and recording a number at that moment in time which represents the height of the sound wave. 

CD quality is usually sampled at 44.1 khz (44 100 'recordings' per second). For speech recognition, a sample rate of 16 kHz does the trick as it is enough to cover the range of the human speech. 

Note: it's not because there are gaps in our data that we are losing information, check out the Nyquist-Shannon theorem (roughly states that we can use our spaced-out samples to reconstruct the original sound wave, given we sample twice as fast the highest frequency we want to record). Also note that using higher quality sampling does not always lead to better audio quality. 

Now we can image we have an array of numbers representing the sound wave's amplitudes at equidistant 1/16000th of a second slices/intervals. We could indeed feed this directly to the computer, although this is regarded as a far more difficult talk and thus we will simplify things by doing some further pre-processing.

## Prepare the data for the model

```python
from __future__ import division, print_function, absolute_import
import sys
import speech_data2
import tensorflow as tf

from keras.models import Sequential
from keras.layers import Dense, Activation, LSTM,Dropout,Conv1D
import numpy as np
from keras import optimizers
```

If your installations went according to plan then you should see 'Using TensorFlow backend.' after running the cell above

Note: running the cell below may take a few minutes on a regular macbook pro or windows. 

```python
## Make sure the data is loaded up and ready to go
Xs,ys = speech_data2.mfcc_batch_generator(2402)


data_size = Xs.shape[0] ## Should be 2402 Xs and 2402 Ys (corresponding to the transcribed digit labels)

## Number of examples we will use for training
Valpercent = 0.1 ## Set up 10 percent of data to validation set 
SPS_per_epoch = int(data_size - data_size*Valpercent)
VALSPS = data_size - SPS_per_epoch ## Number of steps needed for validation data 
```

```python
## Time to split training/validation data 
X_train,y_train = Xs[:SPS_per_epoch,:,:],ys[:SPS_per_epoch,:]
X_val, y_val = Xs[SPS_per_epoch:,:,:],ys[SPS_per_epoch:,:]
```

```python
print('Xs shape: ',Xs.shape)
print('Ys (labels) shape: ', ys.shape)
print('One label for each MFCC') 
```

# Creating the model

```python
## parameters for input,output
width = 20  # mfcc features
height = 80  # (max) length of utterance
classes = 10  # 10 digits 0 through to 9 
```

Well use a recurrent neural networkthat is, a neural network that has a memory that influences future predictions. Thats because each letter it predicts should affect the likelihood of the next letter it will predict too. For example, if we have said HEL so far, its very likely we will say LO next to finish out the word Hello. Its much less likely that we will say something unpronounceable next like XYZ. So having that memory of previous predictions helps the neural network make more accurate predictions going forward.


Before building a RNN, let's go for a simple dense network to warm up and familiarize ourselves with Keras

## Model 0 : fully connected neural network

Build a feedforward neural network with Keras 

```python
## Start with one epoch
NUMEPOCHS = 1
```

Since dense networks on Keras require an integer input, we must flatten the input so that dimensions match

```python
X_train_dense=np.reshape(X_train,[SPS_per_epoch,-1])
X_val_dense = np.reshape(X_val,[VALSPS,-1])
```

```python
print('Build model...')
model = Sequential()
# Keras' convention is that the batch dimension (number of examples) is typically omitted in the input_shape arguments. 
# The batching (number of examples per batch) is handled in the fit call.
model.add(Dense(32,activation='relu',input_dim=np.prod([width,height])))
model.add(Dense(classes,activation='softmax'))
model.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])

for iters in range(NUMEPOCHS):
	print('Epoch #: ',iters+1)
	dotheshuffle = np.random.permutation(SPS_per_epoch)
	keepdoingtheshuffle = np.random.permutation(VALSPS)
	X_train_dense,y_train = X_train_dense[dotheshuffle,:],y_train[dotheshuffle,:]
	X_val_dense,y_val = X_val_dense[keepdoingtheshuffle,:],y_val[keepdoingtheshuffle,:]
	model.fit(X_train_dense[dotheshuffle,:], y_train[dotheshuffle,:], batch_size=None,epochs=1,
		  verbose=1,steps_per_epoch=SPS_per_epoch)
	# score, acc = model.evaluate( was x_test,y_test here
	#                             batch_size=batch_size)
	_,acc = model.evaluate(X_val_dense,y_val)
	# print('Test score:', score)
	print('validation accuracy:', acc)
```

## Model 1: LSTM + fully connected 

```python
## Start with one epoch
NUMEPOCHS = 1
```

```python
print('Build model...')
model = Sequential()
# Keras' convention is that the batch dimension (number of examples) is typically omitted in the input_shape arguments. 
# The batching (number of examples per batch) is handled in the fit call.
model.add(LSTM(16, input_shape=[width,height],recurrent_dropout=0.7))
model.add(Dropout(0.4))
model.add(Dense(classes,activation='softmax'))
model.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])


for iters in range(NUMEPOCHS):
	print('Epoch #: ',iters+1)
	dotheshuffle = np.random.permutation(SPS_per_epoch)
	keepdoingtheshuffle = np.random.permutation(VALSPS)
	X_train,y_train = X_train[dotheshuffle,:,:],y_train[dotheshuffle,:]
	X_val,y_val = X_val[keepdoingtheshuffle,:,:],y_val[keepdoingtheshuffle,:]
	model.fit(X_train[dotheshuffle,:,:], y_train[dotheshuffle,:], batch_size=None,epochs=1,
		  verbose=1,steps_per_epoch=SPS_per_epoch)
	# score, acc = model.evaluate( was x_test,y_test here
	#                             batch_size=batch_size)
	_,acc = model.evaluate(X_val,y_val)
	# print('Test score:', score)
	print('Validation accuracy:', acc)
```

## Model 2: CNN + LSTM + fully connected

Parameters needed for CNN:

```python
## Start with one epoch
NUMEPOCHS = 1
filters = 16
kernel_size= 5
conv_stride = 2
paddingmode = 'SAME'
```

```python
print('Building model...')
model = Sequential()
# Keras' convention is that the batch dimension (number of examples) is typically omitted in the input_shape arguments. 
# The batching (number of examples per batch) is handled in the fit call.
model.add(Conv1D(filters,kernel_size,strides=conv_stride,padding=paddingmode,activation='relu',input_shape=[width,height]))
model.add(LSTM(16,recurrent_dropout=0.7))
model.add(Dropout(0.4))
model.add(Dense(classes,activation='softmax'))
model.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])


for iters in range(NUMEPOCHS):
	print('Epoch #: ',iters+1)
	dotheshuffle = np.random.permutation(SPS_per_epoch)
	keepdoingtheshuffle = np.random.permutation(VALSPS)
	X_train,y_train = X_train[dotheshuffle,:,:],y_train[dotheshuffle,:]
	X_val,y_val = X_val[keepdoingtheshuffle,:,:],y_val[keepdoingtheshuffle,:]
	model.fit(X_train[dotheshuffle,:,:], y_train[dotheshuffle,:], batch_size=None,epochs=1,
		  verbose=1,steps_per_epoch=SPS_per_epoch)
	# score, acc = model.evaluate( was x_test,y_test here
	#                             batch_size=batch_size)
	_,acc = model.evaluate(X_val,y_val)
	# print('Test score:', score)
	print('Validation accuracy:', acc)
```

# Can I build my own speech recognition system?

To build a voice recognition system that performs on the level of Siri, Google Now!, or Alexa, you will need a lot of training datafar more data than you can likely get without hiring hundreds of people to record it for you. And since users have low tolerance for poor quality voice recognition systems, you cant skimp on this. No one wants a voice recognition system that works 80% of the time.

We should keep in mind that recognizing speech is a hard problem. We have to overcome countless challenges: bad quality microphones, background noise, reverb and echo, accent variations, and so on and so forth. All of these issues need to be present in your training data to make sure the neural network can deal with them.
